```python
def generate_walk(G, node, walk_len):
    A = nx.adjacency_matrix(G, weight='None').toarray()
    D = np.diag(list(dict(G.degree()).values()))
    P = np.linalg.inv(D) @ A
    p_0 = len(G.nodes())*[0]
    p_0[node]=1
    eps=0.00001
    i=1
    while np.linalg.norm(np.linalg.matrix_power(P.T, i) - (P.T)@np.linalg.matrix_power(P.T, i))>=eps:
        P=np.linalg.matrix_power(P.T, i)
        i+=1  
    P_inf=np.linalg.matrix_power(P.T, i)
    walk=np.argmax(P_inf.T, axis=1)
    
    
    return walk
```
